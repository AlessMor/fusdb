# Steady-state plasma composition examples

This notebook evaluates steady-state composition for DT, DD, and D-He3 cases using `steady_state_plasma_composition()`.
All available `sigmav_*` inputs are populated for each case.


```python
import numpy as np

from fusdb.relations.plasma_composition.plasma_composition import steady_state_plasma_composition
from fusdb.relations.reactivities.reactivity_functions import (
    sigmav_DT_BoschHale,
    sigmav_DDn_BoschHale,
    sigmav_DDp_BoschHale,
    sigmav_DHe3_BoschHale,
    sigmav_TT_CF88,
    sigmav_He3He3_CF88,
    sigmav_THe3_D_CF88,
    sigmav_THe3_np_CF88,
)
```

```python
constant_profile = np.full(51, 1, dtype=float)
```

# DT case

```python
T_i_DT = 14 * constant_profile# keV
tau_DT = 1 # s
n_i_DT = 1e20 * constant_profile # m^-3

steady_state_DT = steady_state_plasma_composition(
        n_D = 0.5 * n_i_DT,
        n_T = 0.5 * n_i_DT,
        n_He3 = 0.0 * n_i_DT,
        n_He4 = 0.0 * n_i_DT,
        sigmav_DT = sigmav_DT_BoschHale(T_i_DT),
        sigmav_DDn = sigmav_DDn_BoschHale(T_i_DT),
        sigmav_DDp =  sigmav_DDp_BoschHale(T_i_DT),
        sigmav_DHe3 = sigmav_DHe3_BoschHale(T_i_DT),
        sigmav_TT = sigmav_TT_CF88(T_i_DT),
        sigmav_He3He3 = sigmav_He3He3_CF88(T_i_DT),
        sigmav_THe3_D = sigmav_THe3_D_CF88(T_i_DT),
        sigmav_THe3_np = sigmav_THe3_np_CF88(T_i_DT),
        tau_p_D = tau_DT,
        tau_p_T = tau_DT,
        tau_p_He3 = tau_DT,
        tau_p_He4 = tau_DT,  
    )

total_density = steady_state_DT["n_D"] + steady_state_DT["n_T"] + steady_state_DT["n_He3"] + steady_state_DT["n_He4"]
f_T_avg_DT = np.mean(steady_state_DT["n_T"] / total_density)
f_D_avg_DT = np.mean(steady_state_DT["n_D"] / total_density)
f_He3_avg_DT = np.mean(steady_state_DT["n_He3"] / total_density)
f_He4_avg_DT = np.mean(steady_state_DT["n_He4"] / total_density)

print(f"Final DT density fractions:\nf_T_avg = {f_T_avg_DT:.3e}\nf_D_avg = {f_D_avg_DT:.3e}\nf_He3_avg = {f_He3_avg_DT:.3e}\nf_He4_avg = {f_He4_avg_DT:.3e}")
print(f"residual = {1-f_T_avg_DT-f_D_avg_DT-f_He3_avg_DT-f_He4_avg_DT}")
```
