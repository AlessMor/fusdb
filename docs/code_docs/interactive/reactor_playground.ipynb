# Reactor Playground

Edit the input cell below, then run the next cell to load a reactor and print selected solved outputs.


```python
from pathlib import Path

from fusdb import Reactor


def project_root() -> Path:
    root = Path.cwd()
    while not (root / "reactors").exists():
        if root == root.parent:
            raise RuntimeError("Could not locate repository root")
        root = root.parent
    return root
```

```python
reactor_id = "ARC_2015"
output_names = ["P_fus", "Q", "tau_E"]
```

```python
root = project_root()
reactor = Reactor.from_yaml(str(root / "reactors" / reactor_id))
reactor.solve()

for name in output_names:
    variable = reactor.variables_dict.get(name)
    value = None if variable is None else variable.current_value
    print(f"{name}: {value}")
```
